```python
#must import pandas and numpy
import pandas as pd
import numpy as np
```

```python
#we needed to call the saved csv file so pandas can read it and upload
df=pd.read_csv("Health_Sleep_Statistics.csv")
```

```python
#need to print the head to familiarize ourself with the data
# will default to print the first 5 rows of each column in the data set if leave "()" open
print(df.head())
```

```python
#I need to know what each feature is called and which are intergers to convert into arrays
df.info()
```

```python
#creating the array for age and the array Calories burned
#I want to see the median, mean, and stardard deviation for "Age" and "Calories Burned". I can not find the median or mean for non-numerical values unless I convert them into integer or float.
A=np.array(df["Age"])
B=np.array(df["Calories Burned"])
print(A)
print(B)
```

```python
#finding the median of Age and the Calories Burned 
#\n creates a new line before whatever I am asking of the command
print("The median  is:", np.median(A))

print("\nThe median calories burned is:", np.median(B))
```

```python
#finding the mean(average) of Age and the Calories Burned 
#again the \n creates a new line before whatever I am asking of the command
print("The mean age is:", np.mean(A))

print("\nThe mean calories burned is:", np.mean(B))
```

```python
#finding the standard deviation of Age and the Calories Burned to see the furthest from the mean
#again the \n creates a new line before whatever I am asking of the command
print("The standard deviation in age is:", np.std(A))

print("\nThe standard deviation of calories burned is:", np.std(B))
```

```python
#showing mathematical operations by using the arrays
print("Elementwise Multiplication Example:", A*B)
print("\nSubrtaction Example:", A-B)
print("\nAddition Example:", A+B)
```

```python
#the arrays for Bedtime and Wake-up Time. We hope to subtract the Wake-up Time from the Bedtime
A=np.array(df["Bedtime"])
B=np.array(df["Wake-up Time"])
print(A)
print(B)
```

```python
#notice the data above, which is not a float or integer. In order to run mathematical operations we need to change the dtype
```

```python
#Stated making the dtype to be datetime which allows subtraction. asked the format to be in Hour:Minute formatting. Because there is no date, it will default to to January 1st 1900. 
df_new = pd.to_datetime(A, format= '%H:%M')
print(df_new)
```

```python
# we could also format the bedtime and wakeup time like the next two lines below but the other formatting is better for subtraction
#df['Bedtime'] = pd.to_datetime(df['Bedtime'], dayfirst= True)
#df['Wake-up Time'] = pd.to_datetime(df['Wake-up Time'], dayfirst= True)

#this is changing the dtypes of Bedtime and Wake-up Time to be datetime dtypes
df['Bedtime'] = pd.to_datetime(df['Bedtime'], format= '%H:%M')
df['Wake-up Time'] = pd.to_datetime(df['Wake-up Time'],format= '%H:%M')

#note the changed Dtypes for the Aspects "Bedtime" and "Wake-up Time"
df.info()
```

```python
#printing the new arrays now that the Dtype for "Bedtime" and "Wake-up Time" is datetime
A=pd.array(df["Bedtime"])
B=pd.array(df["Wake-up Time"])
print(A)

print(B)
```

```python
#Must subtract the "Wake-Up Time(B)" from the "Bedtime(A)" to get the hours of sleep. 
print("Hours of Sleep:\n", B-A)
```
