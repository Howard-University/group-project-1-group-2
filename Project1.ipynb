## Grouping by gender

```python
grouped_data_by_gender = df.groupby('Gender').agg({
    'Sleep Quality': ['sum', 'mean', 'count', 'min', 'max','std'],
    'Daily Steps': ['sum', 'mean', 'count', 'min', 'max','std'],
    'Calories Burned': ['sum', 'mean', 'count', 'min', 'max','std']
})
grouped_data_by_gender
```

## Grouping by Age Range

```python
# Create age bins (adjust the bin edges as needed)
bins = [18, 30, 40, 50, 60]  # Define the age ranges
labels = ['19-30', '31-40', '41-50', '51-60']  # Corresponding labels

# Create a new column for age range
df['Age Range'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
grouped_data_by_age_range = df.groupby('Age Range', observed=False).agg({
    'Sleep Quality': ['sum', 'mean', 'count', 'min', 'max', 'std'],
    'Daily Steps': ['sum', 'mean', 'count', 'min', 'max', 'std'],
    'Calories Burned': ['sum', 'mean', 'count', 'min', 'max', 'std']
})

# Display the grouped data by age range
grouped_data_by_age_range
```
